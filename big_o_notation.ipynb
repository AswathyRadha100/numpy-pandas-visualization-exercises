```python
plt.figure(figsize=(16, 9))
plt.title('Big O Notation')

plt.plot(x1, y1, label='$y=\sqrt{x}$', color='blue', alpha=0.7)

x1 = list(range(0, 25))
y1 = [0 * n + 1 for n in x1]
plt.plot(x1, y1, label='$O(1)$')

x2 = list(range(1, 10))
y2 = [math.log(n) for n in x2]
plt.plot(x2, y2, label='$O(log n)$')

x3 = list(range(1, 10))
y3 = [n * math.log(n) for n in x3]
plt.plot(x3, y3, label='$O(n)$')

x4 = list(range(0, 10))
y4 = [n**2 for n in x4]
plt.plot(x4, y4, label='$O(n^2)$')

x5 = list(range(0, 8))
y5 = [2**n for n in x5]
plt.plot(x5, y5, label='$O(2^n)$')

x6 = list(range(0, 15))
y6 = [n for n in x6]
plt.plot(x6, y6, label='$O(n log n)$')

x7 = list(range(1, 6))
y7 = [math.factorial(n) for n in x7]
plt.plot(x7, y7, label='$O({n!})$')

x8 = list(range(0, 5))
y8 = [n**n for n in x8]
plt.plot(x8, y8, label='$O(n^n)$')

plt.legend()
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.show()
```

```python
#or

x= (range(1,20))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4= [(n * math.log(n)) for n in x]
y5= [n**2 for n in x]
y6= [2**n for n in x]
y7= [math.factorial(n) for n in x]
y8= [n**n for n in x]

plt.figure(figsize = (16,8))

plt.plot(x,y1, color = 'cyan', label= 'y1= $O(1)$', alpha = .2)
plt.plot(x,y2, color = 'red', label= 'y2 = $O(log n)$', alpha = .3)
plt.plot(x,y3, color = 'b', label= 'y3= $O(n)$', alpha = .4)
plt.plot(x,y4, color= 'g', label= 'y4 = $O(n log n)$', alpha = .5)
plt.plot(x,y5, color= 'coral', label= 'y5 = $O(n^2)$', alpha = .6)
plt.plot(x,y6, color= 'olive', label= 'y6 = $O(2^n)$', alpha = .7)
plt.plot(x,y7, color= 'lime', label= 'y7 = $O(n!)$', alpha = .8)
plt.plot(x,y8, color= 'violet', label= 'y8 = $O(n^n)$', alpha = .9)

plt.title('Big O Notation', fontsize = 20)
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend(loc = 'upper right', fontsize = 14)
plt.xlim(1, 11)
plt.ylim(0, 50)

plt.show()
```
