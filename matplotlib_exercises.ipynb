# Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. There are two common approaches to plotting with Matplotlib.
1)The pyplot API - pyplot is mainly intended for interactive plots and simple cases of programmatic plot generation.
2)The object-oriented API.

```python
Pyplot vs OOP Matplotlib

plt.xlabel() → ax.set_xlabel()
plt.ylabel() → ax.set_ylabel()

plt.xlim() → ax.set_xlim()

plt.ylim() → ax.set_ylim()

plt.title() → ax.set_title()
```

### 1.Use matplotlib to plot the following equation:

y=x^2 − x + 2
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
#a line plot

x = list(range(-50, 52))
y = [n**2-n+2 for n in x]
plt.plot(x,y)
plt.title('A Quadratic Distribution, $y = x^2-x+2$')
plt.xlabel('$x^2-x+2$')
plt.ylabel('$y$')
plt.annotate('origin', xy=(0, 0), xytext=(-4, 500),
             arrowprops= {})
plt.show()
```

```python
#or
#a scatter plot
x = list(range(-49, 50))
y = [n**2-n+2 for n in x]
plt.scatter(x, y)
plt.title('A Quadratic Distribution, $y = x^2-x+2$')
plt.xlabel('$x^2-x+2$')
plt.ylabel('$y$')
plt.annotate('origin', xy=(0, 0), xytext=(0, 1000),
             arrowprops= {'arrowstyle': '->'})
plt.show()
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

You can use functions from the math module to help implement some of the equations above.

```python
#1. y=√x

x1 = list(range(0, 25))
y1 = [n**.5 for n in x1]
plt.title('Chart 1:, $y =\sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x1, y1)
plt.show()
```

### 3.Combine the figures you created in the last step into one large figure with 4 subplots.

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python


plt.figure(figsize=(12, 10))


x1 = list(range(0, 10))
y1 = [n**.5 for n in x1]
plt.plot(x1, y1, label='$y=\sqrt{x}$', color='blue', alpha=0.3)

x2 = list(range(0, 10))
y2 = [n**3 for n in x2]
plt.plot(x2, y2, label='$y=x^3$', color='green', alpha=0.4)

x3 = list(range(0, 10))
y3 = [2**n for n in x3]
plt.plot(x3, y3, label='$y=2^x$', color='red', alpha=0.5)

x4 = list(range(0, 10))
y4 = [1 / (n + 1) for n in x4]
plt.plot(x4,y4,label='$y = 1/(x+1)$', color='yellow', alpha=0.7)

plt.title('Four charts combined, y=√x,y=x^3, y=2^x, y=1/(x+1)')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
# Set the limits for the x and y axes,only the data points within the specified range will be visible on the plot

plt.figure(figsize=(12, 10))


x1 = list(range(0, 10))
y1 = [n**.5 for n in x1]
plt.plot(x1, y1, label='$y=\sqrt{x}$', color='blue', alpha=0.3)

x2 = list(range(0, 10))
y2 = [n**3 for n in x2]
plt.plot(x2, y2, label='$y=x^3$', color='green', alpha=0.4)

x3 = list(range(0, 10))
y3 = [2**n for n in x3]
plt.plot(x3, y3, label='$y=2^x$', color='red', alpha=0.5)

x4 = list(range(0, 10))
y4 = [1 / (n + 1) for n in x4]
plt.plot(x4,y4,label='$y = 1/(x+1)$', color='yellow', alpha=0.7)

plt.title('Four charts combined, y=√x,y=x^3, y=2^x, y=1/(x+1)')
plt.xlabel('x axis')
plt.ylabel('y axis')

#add legend location
plt.legend(loc = 'upper center')


#add gridlines
plt.grid(True, ls= '--')

# Set the limits for the x and y axes
plt.ylim(0,10)
plt.xlim(0,10)

plt.show()

```
