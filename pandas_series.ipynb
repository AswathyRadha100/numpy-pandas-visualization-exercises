#### 4. Write the code necessary to implement a curve for your exam_grades Series and save this as curved_grades. Add the necessary points to the highest grade to make it 100, and add the same number of points to every other score in the Series as well.


```python
#Curving a grade is done by getting the max high score (100) and subtracting the actual highest score. 
#This difference will be added to each number in the Series

# Implement a curve for the exam_grades Series
curve_amount = 100 - maximum_score
curved_grades = exam_scores + curve_amount

print(curved_grades)
```

#### 5. Use a method to convert each of the numeric values in the curved_grades Series into a categorical value of letter grades. For example, 86 should be a 'B' and 95 should be an 'A'. Save this as a Series named letter_grades.


```python
# Convert each numeric value in the curved_grades Series into a categorical letter grade
def get_letter_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

letter_grades = curved_grades.apply(get_letter_grade)
print(letter_grades)
```

```python
# or 
                                                                 
bin_edges = [0, 70, 75, 80, 90, 100]
bin_labels = ['F', 'D', 'C', 'B', 'A']
letter_grades = pd.cut(curved_grades, bins=bin_edges, labels=bin_labels)   
letter_grades
```

```python
#get a count of each grade and reassign the variable
letter_grades = letter_grades.value_counts()
letter_grades
```

#### 6. Plot your new categorical letter_grades Series in a meaninful way and include a title and axis labels.


```python
# Plot the new categorical letter_grades Series
plt.figure(figsize=(5, 3))
letter_grades.value_counts().sort_index().plot(kind='bar', color='y')
plt.xlabel('Letter Grades')
plt.ylabel('Number of Students')
plt.title('Letter Grades Distribution')
plt.show()
```
