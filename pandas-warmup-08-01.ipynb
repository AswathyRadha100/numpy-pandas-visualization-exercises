```python
import pandas as pd


def calculate_stats_difference(dataframe, stat1, stat2):
    # Step 3: Calculate the difference between the two given stats
    dataframe['stats-diff'] = dataframe[stat1] - dataframe[stat2]
    return dataframe

# Example: Calculate difference between Attack and Defense and create a new column 'stats-diff'
df = calculate_stats_difference(df, 'attack', 'defense')
print("\nDataFrame with 'stats-diff' column:")
print(df)



```
